# Encoding the categorical dependent variable

```python
X = data.drop(['Species'], 1)
y = data['Species']
le = LabelEncoder()
y = le.fit_transform(y)
```

# Spliting the dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state = 100)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Building Decision Tree Model

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)
```

```python
clf
```

```python
#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
data_frame = pd.DataFrame({"Actual Data" :y_test, "Predicted Data":y_pred})
```

```python
data_frame.head()
```

```python
data_frame.tail()
```

# Accuracy of the model

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",accuracy_score(y_test, y_pred))
```

# Visualizing the Decision Tree

```python
col_names = ["Sepal length", "Sepal width", "Petal length", "Petal width"]
target_names = ["Setosa", "Versicolor", "Virginica"]
```

```python
plot_tree(clf, feature_names=col_names, class_names=target_names, filled=True)
```
